### Настало время писать код!

В KNN нам нужно для каждого тестового примера найти расстояния до всех точек обучающей выборки. Допустим у нас 1000 примеров в train'е и 100 в test'е, тогда в итоге мы бы хотели получить матрицу попарных расстояний (например, размерностью 100 на 1000). Это можно сделать несколькими способами, и кому-то наверняка, в голову приходит идея с двумя вложенными циклами (надеюсь, что не больше:). Так можно делать, то можно и эффективнее. Вообще, в реальном KNN используется структура данных [k-d-tree](https://ru.wikipedia.org/wiki/K-d-%D0%B4%D0%B5%D1%80%D0%B5%D0%B2%D0%BE), которая позволяет производить поиск за log(N), а не за N, как будем делать мы (по сути это такое расширение бинарного поиска на многомерное пространство).

Вам нужно будет последовательно реализовать методы `compute_distances_two_loops`, `compute_distances_one_loop` и `compute_distances_no_loops` класса `KNN` в файле `knn.py`.

Эти функции строят массив расстояний между всеми векторами в тестовом наборе и в тренировочном наборе.  
В результате они должны построить массив размера `(num_test, num_train)`, где координата `[i][j]` соотвествует расстоянию между i-м вектором в test (`test[i]`) и j-м вектором в train (`train[j]`).

**Обратите внимание** Для простоты реализации мы будем использовать в качестве расстояния меру L1 (ее еще называют [Manhattan distance](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D1%81%D1%82%D0%BE%D1%8F%D0%BD%D0%B8%D0%B5_%D0%B3%D0%BE%D1%80%D0%BE%D0%B4%D1%81%D0%BA%D0%B8%D1%85_%D0%BA%D0%B2%D0%B0%D1%80%D1%82%D0%B0%D0%BB%D0%BE%D0%B2)).

$d_{1}(\mathbf {p} ,\mathbf {q} )=\|\mathbf {p} -\mathbf {q} \|_{1}=\sum _{i=1}^{n}|p_{i}-q_{i}|$

В начале я буду иногда писать разные assert'ы, чтобы можно было проверить правильность реализации, в дальнейшем вам нужно будет их писать самим, если нужно будет проверять корректность каких-то вычислений.

```python
# TODO: compute_distances_two_loops
dists = knn_classifier.compute_distances_two_loops(binary_test_X)
assert np.isclose(dists[0, 100], np.sum(np.abs(binary_test_X[0] - binary_train_X[100])))
```

```python
# TODO: compute_distances_one_loop
dists = knn_classifier.compute_distances_one_loop(binary_test_X)
assert np.isclose(dists[0, 100], np.sum(np.abs(binary_test_X[0] - binary_train_X[100])))
```

```python
# TODO: compute_distances_no_loops
dists = knn_classifier.compute_distances_no_loops(binary_test_X)
assert np.isclose(dists[0, 100], np.sum(np.abs(binary_test_X[0] - binary_train_X[100])))
```

### Подбор оптимального k

Чтобы подобрать оптимальное значение параметра k можно сделать следующее: задать область допустимых значений k, например, `[1, 3, 5, 10]`. Дальше для каждого k обучить модель на тренировочных данных, сделать предсказания на тестовых и посчитать какую-нибудь метрику (метрику выберите сами исходя из задачи, но постарайтесь обосновать выбор). В конце нужно посмотреть на зависимость метрики на train'е и test'е от k и выбрать подходящее значение.

Реализуйте функцию `choose_best_k` прямо в ноутбуке.

```python
# добавила аргумент classifier для определения нужного классификатора (с целью переиспользования функции для KNeighborsRegressor)
def find_best_k(X_train, y_train, X_test, y_test, params, metric, classifier=KNNClassifier):
    """
    Choose the best k for KKNClassifier
    Arguments:
    X_train, np array (num_train_samples, num_features) - train data
    y_train, np array (num_train_samples) - train labels
    X_test, np array (num_test_samples, num_features) - test data
    y_test, np array (num_test_samples) - test labels
    params, list of hyperparameters for KNN, here it is list of k values
    metric, function for metric calculation
    classifier, classifier class
    Returns:
    train_metrics, the list of metric values on train data set for each k in params
    test_metrics, the list of metric values on test data set for each k in params
    """
    
    train_metrics = []
    test_metrics = []
    
    for k in params:
        if classifier == KNNClassifier:
            knn_classifier = classifier(k=k)
        else:
            knn_classifier = classifier(n_neighbors=k)
        knn_classifier.fit(X_train, y_train)
        train_prediction = knn_classifier.predict(X_train)
        test_prediction = knn_classifier.predict(X_test)
        train_metrics.append(metric(y_train, train_prediction))
        test_metrics.append(metric(y_test, test_prediction))
    return train_metrics, test_metrics
```

```python
"""
В качестве метрики была выбрана accuracy_score, т.к. наблюдается допустимый баланс классов. 
Accuracy легче интерпретировать, чем другие метрики; в данном случае графики всех четырех показателей приблизительно схожи, 
а также, скорее всего, в ходе курса мы быстро перейдем от нее к более интересным и сбалансированным показателям, 
посмотрим на нее хотя бы сейчас :)
"""
params = [1, 2, 4, 5, 8, 10, 30, 40, 50]
train_metrics, test_metrics = find_best_k(binary_train_X, binary_train_y, binary_test_X, binary_test_y, params, accuracy_score)
```

```python
plt.plot(params, train_metrics, label="train")
plt.plot(params, test_metrics, label="test")
plt.legend()
plt.xlabel("K in KNN")
plt.ylabel("Accuracy");
```

На самом деле, это не самый лучший способ подбирать гиперпараметры, но способы получше мы рассмотрим в следующий раз, а пока что выберите оптимальное значение k, сделайте предсказания и посмотрите, насколько хорошо ваша модель предсказывает каждый из классов.

В данном случае на графике видно, что качество предсказания на train'е падает с показателя 1.0 при k > 2, на test'е - при k > 40, соответственно, оптимальным будет значение k in [1, 2].

### 2.3. Метрики регресии (3 балла)

Реализуйте метрики $R^2$, MSE и  MAE в `metrics.py`. Примените их для оценки качества полученной модели. Все ли хорошо?

Напомню, что:

$R^2 = 1 - \frac{\sum_i^n{(y_i - \hat{y_i})^2}}{\sum_i^n{(y_i - \overline{y})^2}}$

$MSE = \frac{1}{n}\sum_i^n{(y_i - \hat{y_i})^2}$

$MAE = \frac{1}{n}\sum_i^n{|y_i - \hat{y_i}|}$

### Реализация с Pipeline и стандартизацией

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
raw_df = load_diabetes(as_frame=True, scaled=False)
raw_df = pd.DataFrame(data=raw_df.data, 
                  columns=raw_df.feature_names)
raw_df['y'] = y
raw_df
```

```python
cat_col = ['sex']

num_cols = ['age', 'bmi', 'bp',
            's1', 's2', 's3', 's4', 's5', 's6']

all_features = cat_col + num_cols

target = raw_df['y'].values
```

```python
X = raw_df[all_features]
y = target
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("scaler", StandardScaler(), num_cols),
    ("ohe", OneHotEncoder(drop="first"), cat_col)
])
```

```python
preprocessor.transformers
```

```python
knn_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("knn", KNeighborsRegressor(n_neighbors=8, n_jobs=16))
])
```

```python
knn_pipeline.fit(X, y)
```

```python
predictions = knn_pipeline.predict(X)
```

```python
print(r2_score(y, predictions) == r_squared(predictions, y))
print(mean_squared_error(y, predictions) == mse(predictions, y))
print(mean_absolute_error(y, predictions) == mae(predictions, y))

print(r2_score(y, predictions))
print(mean_squared_error(y, predictions))
print(mean_absolute_error(y, predictions))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=SEED)
```

```python
params = [1, 2, 4, 5, 8, 10, 30]
train_metrics_r_squared, test_metrics_r_squared = find_best_k(X_train, y_train, X_test, 
                                                              y_test, params, 
                                                              r2_score, KNeighborsRegressor)
train_metrics_mse, test_metrics_mse = find_best_k(X_train, y_train, 
                                                  X_test, y_test, 
                                                  params, mean_squared_error, KNeighborsRegressor)
train_metrics_mae, test_metrics_mae = find_best_k(X_train, y_train, 
                                                  X_test, y_test, 
                                                  params, mean_absolute_error, KNeighborsRegressor)
```

```python
plt.suptitle("Dependence of metric values for KNeighborsRegressor on n_neighbors parameters")

plt.subplot(311)
plt.plot(params, train_metrics_r_squared, label="train")
plt.plot(params, test_metrics_r_squared, label="test")
plt.xlabel("K in KNN")
plt.ylabel("r squared")
plt.legend()

plt.subplot(312)
plt.plot(params, train_metrics_mse, label="train")
plt.plot(params, test_metrics_mse, label="test")
plt.xlabel("K in KNN")
plt.ylabel("MSE")
plt.legend()

plt.subplot(313)
plt.plot(params, train_metrics_mae, label="train")
plt.plot(params, test_metrics_mae, label="test")
plt.xlabel("K in KNN")
plt.ylabel("MAE")
plt.legend();
```

```python
n_neighbors = [1, 3, 5, 8, 10, 15, 20, 25, 30]
for k in n_neighbors:
    knn_pipeline = Pipeline(steps=[
                                ("preprocessor", preprocessor),
                                ("knn_reg", KNeighborsRegressor(n_neighbors=k))
                            ])
    knn_pipeline.fit(X_train, y_train)
    y_pred = knn_pipeline.predict(X_test)
    print(f"With k = {k} R2 = {r2_score(y_test, y_pred)}")
    print(f"With k = {k} MSE = {mean_squared_error(y_test, y_pred)}")
    print(f"With k = {k} MAE = {mean_absolute_error(y_test, y_pred)}")
    print()
    
```
